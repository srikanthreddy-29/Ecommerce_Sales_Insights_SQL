## üìä E-Commerce EDA Analysis and Visualizations

This notebook performs **Exploratory Data Analysis (EDA)** on the cleaned e-commerce datasets to uncover patterns, trends, and insights that drive business decisions.


```python
# üì¶ Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings

warnings.filterwarnings("ignore")
sns.set(style="whitegrid")

# üìÇ Load cleaned datasets
cleaned_path = "../datasets/cleaned_data"

customers = pd.read_csv(os.path.join(cleaned_path, "cleaned_customers.csv"))
geolocation = pd.read_csv(os.path.join(cleaned_path, "cleaned_geolocation.csv"))
order_items = pd.read_csv(os.path.join(cleaned_path, "cleaned_order_items.csv"))
orders = pd.read_csv(os.path.join(cleaned_path, "cleaned_orders.csv"))
payments = pd.read_csv(os.path.join(cleaned_path, "cleaned_payments.csv"))
products = pd.read_csv(os.path.join(cleaned_path, "cleaned_products.csv"))
sellers = pd.read_csv(os.path.join(cleaned_path, "cleaned_sellers.csv"))
```

### üîç Basic Info and Null Values Check
Begin with basic info, descriptive statistics, and null checks on all datasets.


```python
datasets = {
    "Customers": customers,
    "Order Items": order_items,
    "Orders": orders,
    "Payments": payments,
    "Products": products,
    "Sellers": sellers
}
```

```python
for name, df in datasets.items():
    print(f"\nüìÑ {name} Dataset".upper())

    print("Data Types and Nulls:")
    display(df.info())

    print("\nüìä Descriptive Statistics:")
    display(df.describe(include='all'))

    print("\nüîç Missing Values:")
    display(df.isnull().sum())
```

### üìà 1. Top 10 Product Categories


```python
plt.figure(figsize=(10,6))
top_categories = products['product category'].value_counts().head(10)
sns.barplot(x=top_categories.values, y=top_categories.index,hue=top_categories.index)
plt.title("Top 10 Product Categories")
plt.xlabel("Count")
plt.ylabel("Product Category")
plt.tight_layout()
plt.show()
```

### üìâ 2. Order Status Distribution


```python
plt.figure(figsize=(8,6))
order_status_counts = orders['order_status'].value_counts()
sns.countplot(data=orders, y='order_status', order=order_status_counts.index)
plt.title("Order Status Distribution")
plt.tight_layout()
plt.show()
```

### üí≥ 3. Payment Type Pie Chart


```python
plt.figure(figsize=(7,8))
payment_types = payments['payment_type'].value_counts()
plt.pie(payment_types.values, labels=payment_types.index, autopct='%1.1f%%', startangle=140)
plt.title("Payment Type Distribution")
plt.tight_layout()
plt.show()
```

### üöö 4. Freight vs Price (Scatter)


```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=order_items, x='price', y='freight_value')
plt.title("Freight vs. Product Price")
plt.tight_layout()
plt.show()
```
